```python
import requests
import os
from urllib.parse import urlparse, parse_qs

def download_file(id, path):
    url = f"https://www.bridgebase.com/tools/vugraph_linfetch.php?id={id}"
    parsed_url = urlparse(url)
    query_params = parse_qs(parsed_url.query)

    file_name = f"{id}.txt"

    if not os.path.exists(path):
        os.makedirs(path)
    file_path = os.path.join(path, file_name)

    with open(file_path, "wb") as f:
        response = requests.get(url)
        f.write(response.content)
    print(f"{file_name} downloaded")

```

```python
import csv
import pandas as pd
import re
def generate_CSV(id, path):
    # 构建文件名
    file_name = f"{id}.txt"
    
    # 构建完整的文件路径
    file_path = os.path.join(path, file_name)
    
    # 读取数据文件
    data = {'Index':[], 'Info':[], 'DEALER':[], 'VUL':[], 'BiddingSequence':[], 'S':[], 'W':[], 'N':[], 'E':[]}
    df = pd.DataFrame(data)
    i = 0
    with open(file_path, 'r') as txt_file:
        text = txt_file.read()
        text = text.replace('\n', '').replace(' ', '').replace('\t', '')
        text = text.lower()
        df.loc[i] = ''
        df.loc[i, 'Index'] =i
        df.loc[0, 'Info'] = re.search('\|(.*?)\|rs', text).group(1) if re.search('\|(.*?)\|rs', text) else ''
        i += 1
        #game information parsed
        for split in re.finditer('\|qx\|.*?\|pc\|', text) : #iter every game in current file
            partition = split.group()
            #print(partition)
            #print(type(partition))
            df.loc[i] = ''
            df.loc[i, 'Index'] =i
            df.loc[i, 'DEALER'] = re.search('\d+.*?(\d)s', partition).group(1) if  re.search('\d+.*?(\d)s', partition) else ''
            df.loc[i, 'VUL'] = re.search('\|sv\|(.)', partition).group(1) if re.search('\|sv\|(.)', partition) else ''
            df.loc[i, 'BiddingSequence'] = re.search('\|mb\|(.*\|p)\|', partition).group(1) if re.search('\|mb\|(.*\|p)\|', partition) else ''
            cards = re.search('\d+.*?\d(s.*)\|sv\|', partition).group(1) if re.search('\d+.*?\d(s.*)\|sv\|', partition) else ''
            #print(cards)
            #print(type(cards))
            hands = re.findall('s[^,\|]*', cards)
            #print(len(re.findall('[Ss][^,\|]*', cards)))
            #print(hands)
            if len(re.findall('s[^,\|]*', cards)) > 3:
                df.loc[i, 'S'] = hands[0]
                df.loc[i, 'W'] = hands[1]
                df.loc[i, 'N'] = hands[2]
                df.loc[i, 'E'] = hands[3]
            i += 1        
            #print(i)
    # 构建输出文件路径
    output_file_name = f"{id}.csv"
    output_file_path = os.path.join(path, output_file_name)
    #print(df)
    
    # 写入数据到 CSV 文件
    df.to_csv(output_file_path, index=False)
    print(f"{output_file_name} formatted")
```

```python
for id in range (9540, 10001):
    path = "./牌局"
    try:
        download_file(id,  path)
    except Exception as e:
        with open('error_log.txt', 'a') as f:
            f.write(str(id))
            f.write(f": An error occurred during dowloading: {e}\n")
        print(f"An error occurred during dowloading: {e}")
    try:
        generate_CSV(id, path)
    except Exception as e:
        with open('error_log.txt', 'a') as f:
            f.write(str(id))
            f.write(f": An error occurred during parsing: {e}\n")
        print(f"An error occurred during parsing: {e}")
        continue
```
